```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dt = pd.read_csv('covid_worldwide.csv')
dt.head()
```

```python
dt.dtypes
```

```python
# dt['Total Cases'] = pd.to_numeric(dt['Total Cases'], errors='coerce', downcast='float')
# dt['Total Deaths'] = pd.to_numeric(dt['Total Deaths'], errors='coerce', downcast='float')
# dt['Total Recovered'] = pd.to_numeric(dt['Total Recovered'], errors='coerce', downcast='float')

dt['Total Cases'] = dt['Total Cases'].str.replace(',', '')
dt["Total Cases"] = dt["Total Cases"].astype(float)
dt["Total Cases"] = dt["Total Cases"] / 1000000

dt['Total Deaths'] = dt['Total Deaths'].str.replace(',', '')
dt["Total Deaths"] = dt["Total Deaths"].astype(float)
dt["Total Deaths"] = dt["Total Deaths"] / 1000000

dt['Total Recovered'] = dt['Total Recovered'].str.replace(',', '')
dt["Total Recovered"] = dt["Total Recovered"].astype(float)
dt["Total Recovered"] = dt["Total Recovered"] / 1000000

dt['Active Cases'] = dt['Active Cases'].str.replace(',', '')
dt["Active Cases"] = dt["Active Cases"].astype(float)
dt["Active Cases"] = dt["Active Cases"] / 1000000

dt['Population'] = dt['Population'].str.replace(',', '')
dt["Population"] = dt["Population"].astype(float)
dt["Population"] = dt["Population"] / 1000000

```
